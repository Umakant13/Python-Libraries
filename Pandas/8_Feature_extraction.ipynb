```python
#Make a new column for episode count
def extractEpisodes(txt):
    check = False

    data = ""
    for i in txt:

        if i == ')':
            check= False
            return data

        if check == True:
            data = data + i
            
        if i == '(':
            check = True
     

    
```

```python
df['Episodes'] = df['Title'].apply(extractEpisodes)
```

```python
df['Episodes'] = df['Episodes'].str.replace(" eps", "")
```

```python
df['Episodes'] = df['Episodes'].astype(int)
df.head()
```

```python
#Make a new column for Duration

def extractDuration(txt):  
    check = False
    data = ""
    for i in range(len(txt)):

        if txt[i] == ')':
            for j in range(i+1, i+20):
                data += txt[j]

            return data
    
```

```python
df['Duration'] = df['Title'].apply(extractDuration)
```

```python
#Extract Months from the Duration

from dateutil.relativedelta import relativedelta
from datetime import datetime

def totalMonths(txt):

    try:
        start, end = txt.split(' - ')

        start_date = datetime.strptime(start, '%b %Y')
        end_date = datetime.strptime(end, '%b %Y')
        r = relativedelta(end_date, start_date)

        return r.years * 12 + r.months + 1

    except:
        return None
```

```python
df['Months'] = df['Duration'].apply(totalMonths)
df.head()
```
