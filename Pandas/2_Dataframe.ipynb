### Creating a DataFrame

```python
data = {
    'Name' : ['Steve Smith', "Stephen Strange", "Peter Parker", "Reed Richards", "Black Panther Shuri", "Killmonger", "Steve Rogers", "Clint Barton"],
    'Age' : [28, 34, 54, 32, 24, 35, 19, 42],
    'City' : ['Sydney', 'New York', 'Paris', 'Berlin', 'Wakanda', 'London', 'New Delhi', 'California'],
    'Salary' : [65000, 20000, 36000, 40000, 50000, 29000, 26000, 37000]
}
```
