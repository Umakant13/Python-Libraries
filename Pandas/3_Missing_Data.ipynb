```python
import pandas as pd
import numpy as np
```

### Finding Missing Data

```python
data = {

    'A' : [1,2, np.nan, 4,5],
    'B' : [np.nan, 2,3,4,5],
    'C' : [1,2,3, np.nan, np.nan],
    'D' : [1, np.nan, np.nan, np.nan, 5],
    'E' : [1,2,3,4,5]
}
```

```python
df= pd.DataFrame(data)
df
```

```python
df.isna()  #To check null values
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

### Removing Missing Data

```python
df
```

```python
df.dropna()  # Drop rows which have null values
```

```python
df.dropna(thresh= 4)  #Thresh = no. of non null values in a row
```

### Filling Missing Data

```python
df
```

```python
df.fillna(0)  #Fill 0 instead of null value
```

```python
df
```

```python
values ={'A': 100, 'B': 100, 'C': 200, 'D':300, 'E': 400} #Different values for each column
df.fillna(value = values)
```

```python
df.fillna(df.mean())
```
