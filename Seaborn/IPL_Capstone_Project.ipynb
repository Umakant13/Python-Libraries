### Loading Libraries and Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('IPL.csv')
df.head()
```

### Basic Questions

#### 1. Which team won the most matches ?

```python
match_wins = df['match_winner'].value_counts()
```

```python
sns.barplot(y = match_wins.index, x = match_wins.values, palette= 'viridis')
plt.title("Team with Most Wins in IPL 2022")
```

#### 2. Toss Decision Trends

```python
sns.countplot(x= df['toss_decision'], palette= 'rainbow')
plt.title("Toss Decision Trends")
```

#### 3. Toss Winner vs Match Winner

```python
df[ df['toss_winner'] == df['match_winner']]
```

```python
count = df[ df['toss_winner'] == df['match_winner']]['match_id'].count()

percentage = (count *100)/ df.shape[0]
percentage.round(2)
```

#### 4. How do teams win ? (Runs vs Wickets)

```python
sns.countplot(x = df['won_by'], palette='viridis')
plt.title("Won By")
```

### Key Player Performances

#### Most "POTM" Awards

```python
potm = df['player_of_the_match'].value_counts().head(15)
potm
```

```python
sns.barplot(y = potm.index, x = potm.values, palette="mako")
plt.title("Players with most POTM awards")
```

#### Top 2 High Scorers

```python
high = df.groupby(by = 'top_scorer')['highscore'].sum().sort_values(ascending= False).head(2)
high
```

```python
high.plot(kind = "barh")
```

#### Top 10 Best Bowling Figures

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
df.head(5)
```

```python
bowling = df.groupby(by = "best_bowling")['highest_wickets'].sum().sort_values(ascending=False).head(15)
bowling
```

```python
sns.barplot(x = bowling.values, y = bowling.index, palette= "viridis")
```

### Custom Questions

#### Team with the Largest Victory Margin (Runs)

```python
df[df['won_by'] == 'Runs'].sort_values(by = 'margin', ascending= False).head(1)[['match_winner', 'margin']]
```

#### Highest Individual Score

```python

high_score = df[df['highscore'] == df['highscore']].sort_values(by = 'highscore', ascending= False).head(10)[['top_scorer', 'highscore']]
high_score
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='highscore', y='top_scorer', data=high_score, palette='rainbow')
plt.xlabel('Runs')
plt.ylabel('Player')
plt.title('Top 10 High Scorers in IPL 2022')
```
