##### Distribution plots are created to visualize the distribution of quantitative data.

##### Some plots that allow us to visualize the distribution of a data set. These plots are

##### 1. Displot / Histplot
##### 2. Jointplot
##### 3. Pairplot
##### 4. Rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
# Displot/ Histplot

plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 20, kde = True)  
plt.subplot(1,2,2)
sns.histplot(df['tip'], kde= True)
```

```python
sns.histplot(df['tip'], bins=20)
```

```python
# Jointplot

sns.jointplot( x = 'total_bill', y= 'tip', data = df, kind = "scatter")
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = df, kind ="kde")
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = "hex")
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
# Pairplot

sns.pairplot(df, hue = 'size', palette= 'rainbow')  #Hue : To differentiate between categories
```

```python
sns.pairplot(df, hue= 'sex')
```

```python
# rugplot

sns.rugplot(df)
```

```python
sns.rugplot(df['total_bill'])
```
