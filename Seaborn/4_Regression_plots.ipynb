## Implot
Regression plot is used in machine learning

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df )
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = df)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = df, palette= 'coolwarm')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = df, palette= 'coolwarm', markers= ['o', 'v'])
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, palette= 'coolwarm', scatter_kws= {'s' : 50})
```
