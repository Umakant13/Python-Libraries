```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Matplotlib Basics

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x **2
y
```

```python
plt.plot(x, y)  #creates line graph
plt.title("Hello World")
plt.xlabel("X- axis")
plt.ylabel("Y- axis")
```

### Subplots in Matplotlib

```python
#Using plt.subplot()

plt.subplot(2,2, 1)  
plt.plot(x**2, y**1/2)

plt.subplot(2,2, 2)
plt.plot(y**2, x)

plt.subplot(2,2, 3)
plt.plot(y, x)

plt.subplot(2,2, 4)
plt.plot(x, y)
```

```python
#Using plt.subplots()
fig, ax = plt.subplots(1,2) #1 row, 2 columns

ax[0].plot(x, y, 'g')
ax[1].plot(y, x)
plt.tight_layout()
```

### Object Oriented Way

```python
figure = plt.figure(figsize=(5, 5), dpi= 500)
axis1 = figure.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)

axis2 = figure.add_axes([0.3, 0.4, 0.3, 0.3])
axis2.plot(x, y)
```

### Types of Plots

```python
plt.scatter(x, y)    # Scatter Plot
```

```python
from random import sample  #Histogram
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
plt.boxplot(x)  #Box Plot -- follows 5 number summary -- Min, 1st Quart, Median, 2nd Quart, Max
```

### Saving Plots

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.5, 0.5])
ax1.plot(x, y)

plt.savefig("Basic Plot.png")
```
